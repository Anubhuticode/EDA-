```python
#missing value
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
for feature in feature_nan:
    print('The feature is {} and missing percntage is {} %'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#categorical feature:
categorical_feature =[feature  for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
data=dataset.copy()
def replace_cat_value(dataset,categorical_feature):
    data[categorical_feature]=data[categorical_feature].fillna('Missing')
    return data
```

```python
replace_cat_value(data,categorical_feature)
```

```python
def replace_cat_value(dataset,categorical_feature):
    data[categorical_feature]=data[categorical_feature].fillna(data[categorical_feature].mode())
    return data
```

```python
replace_cat_value(data,categorical_feature)
```

```python
numerical_feature_nan=[feature for feature in dataset.columns if dataset[feature].dtype!='O' and
                      dataset[feature].isnull().sum()>=1]
```

```python
numerical_feature_nan
```

```python
for feature in numerical_feature_nan:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull,0,1)
    dataset[feature].fillna(median_value,inplace=True)
```
